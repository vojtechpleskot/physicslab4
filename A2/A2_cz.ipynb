## Charakteristika plynového detektoru

Závislost ionizačního proudu na napětí mezi elektrodami detektoru, tj. charakteristika komory, je
důležitá pro výběr způsobu a podmínek provozu detektoru. V normálních případech při
konstantní intenzitě ionizujícího záření vykazuje typickou závislost, kterou lze rozdělit do následujících
oblastí:

- Oblast Ohmova zákona. If no voltage is applied to the detector electrodes,
no electric current flows through the chamber. The ions and electrons created by radiation completely
recombine or diffuse outside the active region of the detector. Applying a small
voltage leads to the separation of free charge carriers of opposite signs. The speed of their
movement to the corresponding electrodes is directly proportional to the intensity of the electric field.
The ionisation current therefore increases in direct proportion to the voltage.
- Se zvyšujícím se napětím jsou nosiče náboje rychleji přitahovány k elektrodám,
počet iontů v aktivní oblasti se snižuje, a tedy se snižuje i pravděpodobnost jejich rekombinace. Úměrnost
proudu vůči napětí přestává platit. Nárůst proudu s napětím se zpomaluje.
- Oblast saturačního proudu je charakterizována konstantní hodnotou ionizačního proudu,
nezávislou na napětí. Při dostatečně vysokém napětí je rychlost rekombinace potlačena
na zanedbatelnou úroveň a veškerý volný náboj vytvořený ionizací přispívá
k ionizačnímu proudu. Další zvýšení napětí nemůže vést ke zvýšení proudu,
protože veškerý volný náboj byl již sebrán a rychlost tvorby náboje je konstantní. Tato
oblast je typická pro provoz ionizačních komor.
- Oblast plynového/lavinového zesílení. Na své cestě k elektrodám prodělávají ionty a elektrony
četné srážky s neutrálními molekulami/atomy plynové náplně detektoru.
Při dostatečně vysoké intenzitě elektrického pole ($10^6~\mathrm{V/m}$ pro typické plyny)
získávají elektrony mezi srážkami energii vyšší než je ionizační energie.
V následné srážce pak mohou vytvořit nový iontový pár. Sekundární iontový pár je
také urychlen elektrickým polem a vytváří další iontové páry. Tímto způsobem
dochází k tzv. lavinovému zesílení, kde každý elektron uvolněný při primární ionizaci
zářením generuje určité množství sekundárních volných elektronů přispívajících
k elektrickému proudu.
- Oblast úměrnosti. V této oblasti je plynové zesílení nezávislé na primární
ionizaci. Ionizace je proto úměrná počtu iontových párů vytvořených přímým
působením záření, tj. také celkové ztrátě energie záření v komoře.
- Oblast omezené úměrnosti. Se zvyšujícím se napětím se ukáže, že
zatímco volné elektrony dosáhnou anody velmi rychle, kladné ionty se ze svého místa vzniku během
stejné doby sotva pohnou a pomalu driftují ke katodě. Pokud
je koncentrace kladných iontů v objemu detektoru dostatečně vysoká, představují
prostorový náboj, který výrazně snižuje intenzitu elektrického pole. To
oslabuje další plynové zesílení a narušuje úměrnost mezi signálem a
primární ionizací. Stále platí, že více energie uložené v detektoru odpovídá
většímu signálu, ale již není přímo úměrné.
- Oblast Geigerova-Müllerova (GM) detektoru. At sufficiently high voltage, the avalanche develops only
until the emerging positive space charge reduces the intensity of the
electric field to a level where gas amplification can no longer occur.
At that point, all detector pulses have the same amplitude, independent of the type and
energy of the radiation.

## Druhy plynových detektorů

 Celá řada přístrojů určených k měření jaderného záření je založena na ionizaci. Různé
charakteristiky popsaného zařízení využívají různé typy plynových čítačů. Oblast
Ohmova zákona není vhodná pro detekci částic, protože výsledné ionizační proudy jsou velmi
malé a jejich velikost je citlivá na přiložené napětí.

 Oblast saturačního proudu I, kde pracují ionizační komory, je pro detekci částic mnohem výhodnější.
 V této oblasti je ionizační proud procházející komorou nezávislý na velikosti
přiloženého napětí a částice s různými energiemi jsou snadno rozlišitelné. Nevýhodou
ionizačních komor je, že ionizační proudy jsou stále velmi malé, což klade nároky na
následné elektronické zpracování, konkrétně na značné lineární zesílení, nízkou úroveň šumu a
podobně.

 Po této oblasti následuje oblast proporcionálních čítačů. Jejich velkou výhodou je
takzvané plynové zesílení, které dosahuje tří až pěti řádů velikosti. Tyto čítače využívají zvýšení
počtu iontů v plynové náplni prostřednictvím sekundární ionizace při srážkách. Protože závislost ionizačního proudu na přiloženém napětí je v této oblasti velmi strmá,
proporcionální čítače vyžadují pro správnou funkci velmi dobře stabilizovaný zdroj napětí.
Na horním okraji oblasti úměrnosti, takzvané oblasti omezené úměrnosti,
charakteristiky komory pro různé typy částic začínají konvergovat a tato část oblasti se
prakticky nevyužívá.

Koncový bod oblasti úměrnosti, kde charakteristiky komory pro různé typy částic
konvergují v jednom bodě, je znám jako Geigerův práh. Geigerovy-Müllerovy čítače pracují nad tímto prahem. Jejich výhodou je vysoké plynové zesílení. Geigerovy-Müllerovy čítače však
nemohou rozlišit mezi energiami nebo druhy částic v dopadajícím záření.

## Ionizační komora

Nejčastěji používané komory jsou statické nízkotlaké komory s konstantní ionizací [1]. Jsou to
v podstatě plynové (vzduchové) kondenzátory. Účinek ionizujícího záření udržuje konstantní ionizační proud, který vybíjí komoru, a velikost ionizačního proudu se
následně měří vhodným způsobem. Pouze u velmi silných zdrojů je možné měřit tento proud vysoce
citlivým galvanometrem. K detekci změny napětí v čase se obvykle používá elektrometr.

Při práci s ionizační komorou lze použít buď pulsní nebo integrální zapojení. V prvním
případě se z odporu připojeného k
nápajecímu obvodu komory snímá buď proud nebo, výhodněji, napěťový puls, který se zesiluje a zaznamenává. Ionizační komora v integrálním zapojení
jako detektor dávky ionizujícího záření. V tomto případě je na kondenzátor reprezentující
ionizační komoru přiloženo napětí $U$ a je na něj přiveden náboj $Q = CU$, kde $C$ je konstantní kapacita
kondenzátoru. Ionizující částice procházejí efektivním objemem komory a zanechávají za sebou
ionty, které jsou přenášeny k elektrodám, kde je jejich náboj neutralizován. To má za následek
přenos náboje mezi elektrodami kondenzátoru a kondenzátor se vybíjí. Protože kapacita
kondenzátoru je v tomto uspořádání konstantní, pokles napětí odpovídá dávce záření
registrované komorou. Časová změna napětí na kondenzátoru pak odpovídá intenzitě
registrovaného záření.

Za normálních podmínek není vzduch v ionizační komoře dokonalým izolátorem. Kvůli tomu
a kvůli kosmickému záření nebo stopovým množstvím radioaktivních prvků, které jsou
přítomny ve všech látkách, se komora vybíjí. Ačkoliv je tento takzvaný svodový proud velmi
malý, je srovnatelný s ionizačním proudem. Proto musí být výsledná měření vždy korigována o
tento efekt.

Ačkoliv jsou ionizační komory v zásadě stejné pro detekci různých typů záření,
liší se v konstrukci. Alfa částice pocházející z přírodně radioaktivních vzorků mají ve vzduchu krátký dosah.
Například dosah nejenergetičtějších alfa částic emitovaných z radioaktivního izotopu
polonia $^{212}\mathrm{Po}$, které mají kinetickou energii $T_0 = 8.776~\mathrm{MeV}$, je za normálních podmínek ve vzduchu menší než $9~\mathrm{cm}$. Proto se alfa zářiče umísťují do válcové komory o
průměru asi $18~\mathrm{cm}$, aby většina alfa částic zanechala svou energii v komoře.

Protože dosah beta částic z přírodně radioaktivních vzorků ve vzduchu za normálních
podmínek činí několik metrů, ionizační komory pro registraci beta záření se konstruují co největší. I tak
se ve efektivním prostoru komory absorbuje pouze zlomek elektronů, a toto zařízení se nejčastěji
používá pro srovnávací měření stejných typů částic se stejnou energií.

Přímá ionizace vzduchu gama zářením je velmi malá. Proto se konstrukce ionizačních komor pro
gama záření značně liší od předchozích dvou zařízení.
Zářič je umístěn mimo komoru a
stěny komory fungují jako konvertor gama záření na elektrony. Protože pravděpodobnost interakce
gama záření s hmotou roste s protonovým číslem látky, používají se komory s tlustými stěnami
z olova, oceli nebo mosazi.

```python
# Solution.

# Fit the first N current-voltage values with a linear function and the last M values with a different linear function.
# Determine the numbers N and M minimizing the chi-squared goodness-of-fit statistic of each linear fit.
# Note that N+M does not need to be equal to the total number of data points - it can be larger or smaller.

# Imports.
from scipy.optimize import curve_fit, OptimizeWarning
import warnings
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import uncertainties

# Define the plotting function.
# Note that 'reg1' and 'reg2' refer to the first and second fitted regions, respectively.
def plot_fits(x, y, y_unc, a_reg1, b_reg1, N_reg1, a_reg2, b_reg2, N_reg2, title = None):

    # Convert inputs to numpy arrays.
    x = np.array(x)
    y = np.array(y)
    y_unc = np.array(y_unc)

    # Plot the data.
    plt.errorbar(x, y, y_unc, fmt='o', label='Data', color='black')

    # Plot the Ohm fit function.
    plt.plot(x[:N_reg1], linear(x[:N_reg1], a_reg1, b_reg1), label='Fit reg. 1', color='red')

    # Plot the saturation fit function.
    plt.plot(x[-N_reg2:], linear(x[-N_reg2:], a_reg2, b_reg2), label='Fit reg. 2', color='blue')

    # get the current legend handles and labels
    handles, labels = plt.gca().get_legend_handles_labels()
    plt.legend(handles=handles[::-1], labels=labels[::-1])  # reverse the order

    # Labels etc.
    plt.xlabel('U [V]')
    plt.ylabel('I [pA]')
    if title:
        plt.title(title)
    plt.show()

# Define the linear model.
def linear(x, a, b):
    return a * x + b

# Function implementing the fit procedure.
# Voltage and current are iterables containing the measured values.
# Uncertainty is an iterable containing the measurement uncertainties for each current value.
# Returns: the fit parameters a, b, and the p-value in the following order: a, b, p-value.
def fit(U, I, I_unc):

    # Fit the linear model to the data.
    try:
        with warnings.catch_warnings():
            warnings.simplefilter("error", OptimizeWarning)
            
            # The curve_fit call is inside the context and try block.
            nom_ab, cov_ab = curve_fit(linear, U, I, sigma=I_unc, absolute_sigma=True)

    except OptimizeWarning as e:
        print(f"Caught an OptimizeWarning message from scipy.optimize.curve_fit: {e}.")
        print(f"The number of points used for fitting was: {len(U)}.")
        return  # Skip to the next iteration if fitting failed

    # Calculate the standard errors of the fitted parameters.
    a, b = uncertainties.correlated_values(nom_ab, cov_ab)

    # Calculate the minimum chi-square value.
    residuals = I - linear(U, *nom_ab)
    chi_square_value = np.sum((residuals / I_unc) ** 2)
    
    # Calculate the degrees of freedom.
    dof = len(U) - len(nom_ab)
    
    # Calculate the p-value.
    p_value = 1 - scipy.stats.chi2.cdf(chi_square_value, dof)

    # Return the fit results.
    return a, b, p_value


for distance in [dist1, dist2]:

    # Get the current and voltage values for the fit.
    if distance == dist1:
        I = df_measurements[f'I_dist1']
    elif distance == dist2:
        I = df_measurements[f'I_dist2']
    else:
        raise ValueError("Unknown distance of the electrodes!")
    U = df_measurements['U']

    # Get the uncertainty.
    I_unc = [uncertainty[distance] for _ in U]

    # Start with the low-voltage region, i.e. the first N points.
    p_reg1 = 0.0
    N_reg1 = 0
    a_reg1, b_reg1 = 0, 0

    # For each fit, determine the chi-squared p-value and compare it with p_reg1.
    for N in range(3, len(U)):

        # Perform the fit.
        try:
            a, b, p_value = fit(U[:N], I[:N], I_unc[:N])
        except:
            print("The fitted region was the region 1.")
            print("Skipping this fit.")
            continue

        # Check if the p-value is larger than the previous maximum.
        if p_value > p_reg1:
            p_reg1 = p_value
            N_reg1 = N
            a_reg1, b_reg1 = a, b

    # Second, do the saturated current region, i.e. the last N points.
    p_reg2 = 0.0
    N_reg2 = 0
    a_reg2, b_reg2 = 0, 0

    # For each fit, determine the chi-squared p-value and compare it with p_reg2.
    for N in range(3, len(U)):

        # Perform the fit.
        try:
            a, b, p_value = fit(U[-N:], I[-N:], I_unc[-N:])
        except:
            print("The fitted region was the region 2.")
            print("Skipping this fit.")
            continue

        # Check if the p-value is larger than the previous maximum.
        if p_value > p_reg2:
            p_reg2 = p_value
            N_reg2 = N
            a_reg2, b_reg2 = a, b


    # Print the results.
    print()
    print(f"Electrodes distance: {distance} cm")
    print("=========================")
    print()
    print("  Ohm region:")
    print("  -----------")
    print(f"  N = {N_reg1}, p-value = {p_reg1:.7f}")
    print(f"  Fit parameters: a = {a_reg1:.4f}, b = {b_reg1:.4f}")
    print()
    print("  Saturated region:")
    print("  -----------------")
    print(f"  N = {N_reg2}, p-value = {p_reg2:.7f}")
    print(f"  Fit parameters: a = {a_reg2:.4f}, b = {b_reg2:.4f}")
    print()

    # Plot the data and the fit.
    plot_fits(U, I, I_unc,
            a_reg1.nominal_value, b_reg1.nominal_value, N_reg1,
            a_reg2.nominal_value, b_reg2.nominal_value, N_reg2,
            title = f"Ionisation chamber, electrode distance: {distance} cm")
```
