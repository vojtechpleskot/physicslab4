# Beta zen a jeho absorpce

Beta zen se skld z elektron emitovanch pi beta rozpadech atomovch jader:
$$
^{A}_{Z}\mathrm{X} \rightarrow ^{A}_{Z+1}\mathrm{Y} + e^- + \bar{\nu}_e,
$$
kde X a Y oznauj matesk a dcein jdro, $e^-$ elektron a $\bar{\nu}_e$ a elektronov antineutrino. Jeliko se jedn o rozpad na ti stice, kinetick energie produkt rozpadu zvis nejen na jejich hmotnostech, ale tak na hlech, pod ktermi od sebe odltaj. Energie emitovanch elektron tak tvo spojit spektrum od nuly do urit maximln hodnoty. Ta se typicky pohybuje v rozmez od destek keV do jednotek MeV v zvislosti na rozpadajcm se izotopu X (mateskm jde). Elektrony emitovan pi beta rozpadu interaguj hlavn excitac nebo ionizac atomovho obalu nebo elastickmi srkami s atomovmi jdry. To je postupn zpomaluje a mn smr jejich letu, take nkter z nich se ji nedostanou do detektoru - tato ztrta se nazv absorpce. Absorpn kivka je pak definovna jako zvislost potu elektron prochzejcch absorbtorem na tlouce absorbtoru. Tvar tto kivky pro monoenergetick elektrony nen exponenciln (narozdl od foton). Kdy ale vezmeme v vahu skutenost, e elektrony z beta rozpadu maj spojit spektrum, a nikoliv monoenergetick, zskme absorpn kivku, jej tvar lze aproximovat exponenciln funkc (toto je netriviln fakt, kter zde nebudeme diskutovat):
$$
N(d) = N(0) e^{-\mu(E_0) d};
$$
$N(d)$ oznauje poet elektron prochzejcch absorbtorem s plonou hustotou $d$. Veliina $\mu$ pedstavuje absorpn koeficient, jeho hodnota zvis na maximln pozorovan energii beta rozpadu $E_0$ a na materilu absorbtoru. Pouit zi, $^{90}\mathrm{Sr}$, se rozpad v kaskd dvou po sob nsledujcch beta rozpad:
$$
^{90}_{38}\mathrm{Sr} \xrightarrow{\beta^-}\ ^{90}_{39}\mathrm{Y} \xrightarrow{\beta^-}\ ^{90}_{40}\mathrm{Zr}
$$
kde se maximln energie elektron emitovanch v kadm kroku kaskdy $E_{0}^{1}$, $E_{0}^{2}$ li. Odpovdajc absorpn kivka je tedy dna soutem dvou exponencilnch funkc
$$
N(d) = N_{1}(0) e^{-\mu(E_{0}^{1}) d} + N_{2}(0) e^{-\mu(E_{0}^{2}) d} + N_\mathrm{B},
$$
kde jsme tak pidali konstantn len $N_B$, kter odpovd pozad signl, jejich frekvence nezvis nebo jen slab zvis na tlouce absorbtoru.

$\color{red}{\textbf{loha:}}$ Zmte absorpn kivku beta zen emitovanho $^{90}\mathrm{Sr}$ pomoc GM detektoru.

- Men se provd s GM detektorem pi napt v oblasti plata.
- Doba men je nastavena na 100 s pro kadou tlouku absorbtoru.
- Hlinkov desky jsou postupn pidvny mezi zdroj a detektor od nejtench po nejtlust. Tmto zpsobem je pokryt interval plonch hustot od 0 do piblin $1100\ \mathrm{mg}\,\mathrm{cm}^{-2}$.
  - Po pidn vech hlinkovch desek se provede jet jedno men s olovnou deskou.

```python
# Solution.

# TODO: Insert your measured data into a DataFrame whose columns are named 'd' (for the absorber thickness), 'N' (for the number of events), and 'N_unc' (for the uncertainty on N).
# Read your measured data from the text file containing it.
# You can check the meaning of the parameters of the function pd.read_csv() in the code above.
# Convert d to g / cm^2.
# Add a column 'N_unc' to the DataFrame that contains the uncertainty on N.

df = pd.read_csv('example_data/absorption.txt', sep='\s+', skiprows=1, usecols=[1, 2], header=None)
# remove the last row
df = df[:-1]
# name the columns
df.columns = ['d', 'N']
# cast the columns to float
df['d'] = df['d'].astype(float) / 1000.
df['N'] = df['N'].astype(float)
# calculate the uncertainty on N
df['N_unc'] = df['N']**0.5
print(df.rename(columns={'d': 'd [g/cm^2]', 'N': 'N', 'N_unc': 'N_unc'}))
```

$\color{red}{\textbf{loha:}}$ Vyneste namen body a fitem naleznte hodnoty parametr absorpn kivky.

```python
# Solution.

import matplotlib.pyplot as plt
from IPython.display import display, Latex
import uncertainties
from uncertainties import unumpy
import numpy as np
from scipy.optimize import curve_fit

# Drawing function definition.
def plot(x, y, y_err, x_fit, y_fit, xlabel, ylabel):

    # draw x, y with error bars
    plt.errorbar(x, y, y_err, fmt='o', label='Data', color='black')

    # draw the fit function and its uncertainty band
    plt.plot(x_fit, y_fit, color='red', alpha=0.3, label='Fit')

    # get the current legend handles and labels and reverse their order
    handles, labels = plt.gca().get_legend_handles_labels()
    plt.legend(handles=handles[::-1], labels=labels[::-1])

    # finally, plot
    plt.plot(x_fit, y_fit, 'r-')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

    return


# Define the absorption function
def absorption_curve(d, N_1, mu_E1, N_2, mu_E2, N_B):
    return N_1 * np.exp(-mu_E1 * d) + N_2 * np.exp(-mu_E2 * d) + N_B

# fit the resolution function to the data
nom, cov = curve_fit(absorption_curve, df['d'], df['N'], sigma = df['N_unc'], absolute_sigma = True)

# to draw the fit function and its uncertainty band, we need to evaluate the fit function at many points
x = np.linspace(df['d'].min(), df['d'].max(), 100)
y_fit = absorption_curve(x, *nom)

# plot the data and the fit function
plot(df['d'], df['N'], df['N_unc'], x, y_fit, '$d$ [g $\mathrm{cm}^{-2}$]', '$N$')

# extract the fit parameters and their uncertainties
N_1, mu_E1, N_2, mu_E2, N_B = uncertainties.correlated_values(nom, cov)

# print the fit parameters
display(Latex(f'$\mu_{{E^1_0}} = {mu_E1}~\mathrm{{g}}^{{-1}}\ \mathrm{{cm}}^2$'))
display(Latex(f'$\mu_{{E^2_0}} = {mu_E2}~\mathrm{{g}}^{{-1}}\ \mathrm{{cm}}^2$'))
display(Latex(f'$N_\mathrm{{B}} = {N_B}$'))
```

$\color{red}{\textbf{loha:}}$ Vypotte maximln energii elektron emitovanch pi beta rozpadu $^{90}\mathrm{Sr}$. Pro zde pouit hlink existuje empirick vztah mezi $\mu$ a $E_0$:
$$
\mu[g^{-1} \mathrm{cm}^2] = 17(E_0[\mathrm{MeV}])^{-1.43},
$$
do kterho je nutn dosadit v uvedench jednotkch.

# Dosah elektron

Dosah elektron pro danou maximln energii, $R_\beta(E_0)$, lze hrub odhadnout jako tlouku, pi kter je poet elektron prochzejcch absorbtorem tak nzk, e je statisticky nerozliiteln od pozad. Tj. odhadnme jej jako tlouku, pi kter je poet prochzejcch elektron roven nejistot veliiny $N_\mathrm{B}$.

$\color{red}{\textbf{loha:}}$ Odhadnte $R_\beta$ pro ob maximln energie elektron z rozpadov ady $^{90}\mathrm{Sr}$.

```python
# Solution.

# Solve the equation N_1 * np.exp(-mu_E1 * R) = N_B.std_dev for R.

import uncertainties.umath

R_1 = -uncertainties.umath.log(N_B.std_dev / N_1) / mu_E1
R_2 = -uncertainties.umath.log(N_B.std_dev / N_2) / mu_E2
display(Latex(f'$R^1_{{\\beta}} = {R_1}\ \mathrm{{g}}\,\mathrm{{cm}}^{{-2}}$'))
display(Latex(f'$R^2_{{\\beta}} = {R_2}\ \mathrm{{g}}\,\mathrm{{cm}}^{{-2}}$'))
```

$\color{red}{\textbf{loha:}}$ Vyneste ob absorpn kivky a odhadovan dosahy elektron.

```python
# # Solution.

# Plot the curves for the two decays in the same plot. Use the same x-axis range for both curves.
# Plot the above estimated values of R_1 and R_2 with vertical lines in the plot.
# The plot should have a logarithmic y-axis.

# x values for the plot.
x = np.linspace(df['d'].min(), df['d'].max(), 100)

# wrap the absorption function to handle uncertainties
wrapped_absorption = uncertainties.wrap(absorption_curve)

# the first absorption curve
y1 = [wrapped_absorption(x, N_1, mu_E1, 0, 0, 0) for x in x]
y1_upper = unumpy.nominal_values(y1) + unumpy.std_devs(y1)
y1_lower = unumpy.nominal_values(y1) - unumpy.std_devs(y1)
plt.plot(x, unumpy.nominal_values(y1), label='$N_{1}(d)$', linewidth=1, color='red')
plt.fill_between(x, y1_lower, y1_upper, color='red', alpha=0.3)

# the second absorption curve
y2 = [wrapped_absorption(x, 0, 0, N_2, mu_E2, 0) for x in x]
y2_upper = unumpy.nominal_values(y2) + unumpy.std_devs(y2)
y2_lower = unumpy.nominal_values(y2) - unumpy.std_devs(y2)
plt.plot(x, unumpy.nominal_values(y2), label='$N_{2}(d)$', linewidth=1, color='blue')
plt.fill_between(x, y2_lower, y2_upper, color='blue', alpha=0.3)

# vertical lines at R_1 and R_2
plt.vlines(x=R_1.nominal_value, ymin=1, ymax=N_B.std_dev, label='$R_1$', linestyle='--', linewidth=1, color='red')
plt.vlines(x=R_2.nominal_value, ymin=1, ymax=N_B.std_dev, label='$R_2$', linestyle='--', linewidth=1, color='blue')

# horizontal line at N_B.std_dev
plt.axhline(y=N_B.std_dev, color='black', linestyle='--')

# labels etc.
plt.yscale('log')
plt.ylim(1, 1e4)
plt.xlabel('$d$ [g $\mathrm{cm}^{-2}$]')
plt.ylabel('$N$')
plt.legend()
plt.show()
```

# Poissonovo rozdlen

Detekce jednotlivch elektron emitovanch pi beta rozpadu v uritm asovm intervalu jsou vzjemn nezvisl nhodn udlosti. Pravdpodobnost $P(k)$, e se takov udlost vyskytne pesn $k$-krt, je dna Poissonovm rozdlenm:
$$
P(k) = \frac{\mu^k}{k!} e^{-\mu},
$$
kde $\mu$ pedstavuje stedn hodnotu. Smrodatn odchylka tohoto rozdlen je $\sqrt{\mu}$. Se zvyujcm se $\mu$ Poissonovo rozdlen konverguje ke Gaussovu rozdlen:
$$
P(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}},
$$
kde $\sigma = \sqrt{\mu}$.

$\color{red}{\textbf{loha:}}$ Ukate, e poet emitovanch elektron je popsn Poissonovm rozdlenm.

- Men se provd bez absorbtoru, 765krt, pokad po dobu 1 s.

```python
# Solution.

# TODO: Insert your measured data into a DataFrame whose only column is named 'N'.
# Read your measured data from the text file containing it.
# You can check the meaning of the parameters of the function pd.read_csv() in the code above.

df = pd.read_csv('example_data/poisson.txt', sep='\s+', skiprows=1, usecols=[2], header=None)
df.columns = ['N']
```

$\color{red}{\textbf{loha:}}$ Prolote data Poissonovm rozdlenm.

- Pamatujte, e stedn hodnota Poissonova rozdlen, $\mu$, se odhaduje jako prmr namench hodnot.
- Smrodatn odchylka $\mu$ se pak odhaduje jako odmocnina variance dlen odmocninou potu men.

```python

# Solution.

# The mean value estimate
mu_fit = df['N'].mean()

# Calculate the standard deviation of the mean
mu_fit_unc = df['N'].std() / np.sqrt(len(df))

# Print the fit parameter
display(Latex(f'$\mu = {mu_fit:.2f} \pm {mu_fit_unc:.2f}$'))
```

$\color{red}{\textbf{loha:}}$ Ovte kvalitu proloen pomoc $\chi^2$ testu - vypotte p-hodnotu a diskutujte, zda je proloen pijateln.

```python
# Solution.

# Calculate the chi-squared value of the fit
residuals = df['N'] - mu_fit
chi_square = np.sum(residuals ** 2 / mu_fit)
ndf = len(df) - 1  # number of degrees of freedom
display(Latex(f'$\chi^2 = {chi_square:.2f}$'))
display(Latex(f'ndf $= {ndf}$'))

# Calculate the p-value.
import scipy
p_value = 1 - scipy.stats.chi2.cdf(chi_square, ndf)
display(Latex(f'$p = {p_value:.4f}$'))
```

$\color{red}{\textbf{loha:}}$ Vytvote histogram vsledk.
- Histogram by ml bt normalizovan na jednotkovou plochu (tj. kad bin by ml bt dlen celkovm potem men).

```python
# Solution.

# histogram the data
hist, bin_edges = np.histogram(df['N'], bins=[df['N'].min() - 0.5 + i for i in range(df['N'].max() - df['N'].min() + 1)], density=True)
```

$\color{red}{\textbf{loha:}}$ Vyneste histogram a proloen Poissonovo rozdlen.

```python
# Solution.

# to draw the fit function and its uncertainty band, we need to evaluate the fit function at many points
x = np.arange(df['N'].min(), df['N'].max())
from scipy.stats import poisson
y_fit = poisson.pmf(x, mu_fit)

# calculate the bin centers
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# plot the data and the fit function
plot(bin_centers, hist, None, x, y_fit, '$N$', 'Probability')
```
